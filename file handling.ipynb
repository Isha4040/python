```python

def add_record(roll, name, marks, filename="students.txt"):
    with open(filename, "a") as file:
        file.write(f"{roll},{name},{marks}\n")
    print("Record added successfully!")



def view_records(filename="students.txt"):
    print("---- Student Records ----")
    try:
        with open(filename, "r") as file:
            for line in file:
                roll, name, marks = line.strip().split(",")
                print(f"Roll: {roll}, Name: {name}, Marks: {marks}")
    except FileNotFoundError:
        print("No records found yet!")
```

```python
add_record("102", "Bob", "95")
add_record("103", "Charlie", "78")
add_record(roll="101", name="Alice", marks="89")

```

```python
view_records()
```

```python
def find_replace(file_name, old_word, new_word):
    try:
        # Read the file content
        with open(file_name, "r") as file:
            content = file.read()

        # Replace the word
        content = content.replace(old_word, new_word)

        # Write back the updated content
        with open(file_name, "w") as file:
            file.write(content)

        print("Word replaced successfully!")

    except FileNotFoundError:
        print("File not found!")
```

```python
find_replace("sample.txt", "hello", "hi")
```

```python
def remove_blank_lines(file_name):
    try:
        # Read all lines
        with open(file_name, "r") as file:
            lines = file.readlines()

        # Keep only non-blank lines
        clean_lines = [line for line in lines if line.strip() != ""]

        # Write cleaned lines back
        with open(file_name, "w") as file:
            file.writelines(clean_lines)

        print("Blank lines removed successfully!")

    except FileNotFoundError:
        print("File not found!")
```

```python
remove_blank_lines("sample.txt")
```

```python
def copy_file(source, destination):
    try:
        with open(source, "r") as src:
            content = src.read()

        with open(destination, "w") as dest:
            dest.write(content)

        print("File copied successfully!")

    except FileNotFoundError:
        print("Source file not found!")
```

```python
copy_file("input.txt", "output.txt")
```

```python
def merge_files(file_list, output_file):
    try:
        with open(output_file, "w") as outfile:
            for filename in file_list:
                with open(filename, "r") as infile:
                    outfile.write(infile.read())
                    outfile.write("\n")   # add a newline between files
        print("Files merged successfully!")

    except FileNotFoundError as e:
        print(f"Error: {e}")
```

```python
files = ["file1.txt", "file2.txt", "file3.txt"]
merge_files(files, "merged_output.txt")
```
